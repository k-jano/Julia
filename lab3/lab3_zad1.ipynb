```julia
Base.Random.randstring 


N=100000
M=10000
counter=100

function huge()
    x=Array{String}(N)
    for i= 1:N
        x[i]=randstring(6)
    end 
end


function small()
    y=Array{String}(M)
    for i= 1:M
        y[i]=randstring(6)
    end
end

function myFun()
    for i= 1:counter
        huge()
        small()
    end   
end    
        
```

```julia
myFun()
```

```julia
Profile.clear()
@profile myFun()

Profile.print()
```
